```python
def tail_fib(N,a,b):
    if N==2:
        return b
    
    return tail_fib(N-1,b,a+b)
```

```python
tail_fib(10,1,1)
```

```python


def hanoi(N,cl='a',aux='b',target='c'):
    if N==1:
        print(f'Plate {N} from {cl} to {target}')
        return
    hanoi(N-1,cl,target,aux)
    print(f'Plate {N} from {cl} to {target}')
    hanoi(N-1,aux,cl,target)
    
    
```

```python
hanoi(10)
```
